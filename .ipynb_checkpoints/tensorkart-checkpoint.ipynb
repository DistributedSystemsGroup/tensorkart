```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import norm
import seaborn as sns
from PIL import Image
from time import time
from skimage.io._plugins.pil_plugin import (pil_to_ndarray, ndarray_to_pil, _palette_is_grayscale)
import os
from skimage.transform import resize
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import img_as_float
from sklearn.decomposition import PCA
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D
from keras import optimizers
from keras import backend as K
```

```python
# SRC_W = 640
# SRC_H = 480
SRC_W = 615
SRC_H = 130
SRC_D = 3

OFFSET_X = 0
OFFSET_Y = 0

IMG_W = 200
IMG_H = 66
# IMG_W = 50
# IMG_H = 16
IMG_D = 3
```

```python
def load_sample(subdirec):
    image_files = np.loadtxt(subdirec + 'data.csv', delimiter=',', dtype=bytes, usecols=(0,)).astype(str) #luigi_raceway1/data.csv
    joystick_values = np.loadtxt(subdirec + 'data.csv', delimiter=',', usecols=(1,)) # default float
    return image_files, joystick_values

def prepare_image(img):
    if str(type(img)) == "<class 'PIL.Image.Image'>":
        img = img.resize((IMG_W, IMG_H)) # 200, 66
    else:
        img = img.reshape(SRC_H, SRC_W, SRC_D)
#         img = img.reshape(SRC_H, SRC_W)
        img = Image.fromarray(img)
        img = img.resize((IMG_W, IMG_H))
    im_arr = pil_to_ndarray(img)
    return img_as_float(im_arr) # in ndarray out ndarray of float64 # shape is unchanged

def prepare(direc):
    X = []
    y = []
    for subdirec in direc:
        print(subdirec)
        image_files, joystick_values = load_sample(subdirec)
        y.append(joystick_values) # list of ndarrays: steering [and throttle]
        # y is a list of num_folders ndarrays, each of which has size (num_samples_in_one_folder, 2) 
        for image_file in image_files:
            image = imread(image_file) # returns ndarray
#             image = rgb2gray(imread(image_file)) # returns ndarray
            vec = prepare_image(image) 
            X.append(vec)
    X = np.asarray(X) # before: list of ndarrays with shape (200, 60)
    y = np.concatenate(y) 
    return X, y
```

```python
train_dirs = [
    "./samples/choco/", 
    "./samples/choco_flipped/", 
    "./samples/kalamari_desert1/", 
    "./samples/kalamari_desert1_flipped/", 
    "./samples/mario_raceway_center1/", 
    "./samples/mario_raceway_center1_flipped/", 
    "./samples/mario_raceway_center4/", 
    "./samples/mario_raceway_center4_flipped/"
             ]
test_dirs = ["./samples/royal_raceway3/", 
             "./samples/royal_raceway3_flipped/"]

X_train, t_train = prepare(train_dirs)
X_test, t_test = prepare(test_dirs)
```

```python
def flat_norm(images):
    x = np.reshape(images, (images.shape[0], -1))
    x /= x.max() 
    return x

def get_design_matrix(samples, order):
    return np.array([[(x_i)**k for k in range(0, order+1)] for x_i in samples]).reshape([samples.shape[0], -1])

def predictBayesianRegression(images, mu, sigma, sigma2=1e-3, order=2):
    def normal(x, mu, sigma2):
        return 1 / (np.sqrt(2*np.pi*sigma2)) * np.exp(-((x-mu)**2)/(2*sigma2))
    
    def normalMultiVariate(x, mu, sigma):
        k = sigma.shape[0]
        print(x.shape)
        print(mu.shape)
        print(sigma.shape)
        return 1/((2*np.pi)**k * det(sigma)) * np.exp( ((x-mu)**2).T.dot(inv(sigma)).dot(x-mu))
    
    t0 = time()
    samples_new = flat_norm(images)
    X_new = get_design_matrix(samples_new, order)
    post_mean = X_new.dot(mu)
    # post_mean ok
    #print(post_mean)
    post_cov = (sigma2 + X_new.dot(inv(sigma)).dot(X_new.T)).diagonal()
    #print(post_cov.diagonal())
    print('Predictions done in %.2f seconds.' % (time() - t0))
    return post_mean, post_cov
    #return normal(x_new, mu=post_mean, sigma2=post_cov)

def get_Sigma(X, sigma2=1e-3):
    S = np.zeros([X.shape[1], X.shape[1]])
    np.fill_diagonal(S, 1/np.random.random(X.shape[1]))
    return np.linalg.inv(X.T.dot(X)/sigma2 + np.linalg.inv(S))
    
def get_mu(X, t, Sigma, sigma2=1e-3):
    return (1/sigma2 * Sigma.dot(X.T).dot(t))
    
def bayes_lin_regr(X_train, t_train, X_test, sigma2=1e-3, order=1):    
    t0 = time()
    X_train = flat_norm(X_train)
    des_mat = get_design_matrix(X_train, order)
    Sigma = get_Sigma(des_mat, sigma2)
    mu = get_mu(des_mat, t_train, Sigma, sigma2)
    print('Training completed in %.2f seconds.' % (time() - t0))
    des_mat_test = get_design_matrix(X_test, order)
    pred_mean = des_mat_test.dot(mu)
    pred_var = (sigma2 + des_mat_test.dot(np.linalg.inv(Sigma)).dot(des_mat_test.T)).diagonal()
    
    return pred_mean, pred_var
```

```python
pred_mean, pred_var = bayes_lin_regr(X_train, t_train, X_train, sigma2=1, order=3)
```

```python
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

pca = PCA(n_components=X_train_flat.shape[1])
pca.fit(X_train_flat)
X_train_pca = pca.transform(X_train_flat)
X_test_pca = pca.transform(X_test_flat)

exvar = (pca.explained_variance_ratio_)
exvar.sort()
exvar = exvar[::-1]
cdf = np.zeros_like(exvar)
for i in range(len(exvar)):
    cdf[i] = exvar[0] if i == 0 else cdf[i-1]+exvar[i]
cdf = np.insert(cdf, 0, 0) 

fig = plt.figure(figsize=[17, 10])
ax = plt.gca()
x_ax = np.arange(0, X_train_flat.shape[1]+1)

f1 = np.where(cdf >= 0.5)[0][0]
ax.axhline(0.5, color='red', linestyle='-.', lw=2, alpha=0.5)
ax.axvline(f1, linestyle='-.', color='red', lw=2, alpha=0.5,)
ax.text(f1+5, 0.1, '%s features' % f1)

f2 = np.where(cdf >= 0.9)[0][0]
ax.axhline(0.9, color='orange', linestyle='-.', lw=2, alpha=0.5)
ax.axvline(f2,  linestyle='-.',color='orange', lw=2, alpha=0.5)
ax.text(f2+5, 0.1, '%s features' % f2)

f3 = np.where(cdf >= 0.99)[0][0]
ax.axhline(0.99, color='green', linestyle='-.',lw=2, alpha=0.5)
ax.axvline(f3, linestyle='-.', color='green', lw=2, alpha=0.5)
ax.text(f3+5, 0.1, '%s features' % f3)

plt.plot(x_ax, cdf, lw=2.5)

plt.xlabel('Number of features in the latent space')
plt.ylabel('Cumulative explained variance')
plt.title('Principal Component Analysis for MNIST')

plt.show()
```

```python
OUT_SHAPE = 1
INPUT_SHAPE = (IMG_H, IMG_W, IMG_D)

def customized_loss(y_true, y_pred, loss='euclidean'):
    if loss == 'L2':
        L2_norm_cost = 0.001
        val = K.mean(K.square((y_pred - y_true)), axis=-1) \
                    + K.sum(K.square(y_pred), axis=-1)/2 * L2_norm_cost
    # euclidean distance loss
    elif loss == 'euclidean':
        val = K.sqrt(K.sum(K.square(y_pred-y_true), axis=-1))
    return val

def create_model(keep_prob = 0.8):
    model = Sequential()

    model.add(Conv2D(24, kernel_size=(5, 5), strides=(2, 2), activation='relu', input_shape= INPUT_SHAPE))
    model.add(Conv2D(36, kernel_size=(5, 5), strides=(2, 2), activation='relu'))
    model.add(Conv2D(48, kernel_size=(5, 5), strides=(2, 2), activation='relu'))
    model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
    model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
    model.add(Flatten())
    model.add(Dense(1164, activation='relu', name="fcl1"))
    drop_out = 1 - keep_prob
    model.add(Dropout(drop_out))
    model.add(Dense(100, activation='relu', name="fcl2"))
    model.add(Dropout(drop_out))
    model.add(Dense(50, activation='relu', name="fcl3"))
    model.add(Dropout(drop_out))
    model.add(Dense(10, activation='relu', name="fc4"))
    model.add(Dropout(drop_out))
    model.add(Dense(OUT_SHAPE, activation='softsign'))

    return model

epochs = 2
batch_size = 50

model = create_model()
model.compile(loss=customized_loss, optimizer=optimizers.adam())
model.fit(X_train, t_train, batch_size=batch_size, epochs=epochs, shuffle=True, validation_split=0.1)

score = model.evaluate(X_test, t_test)
# print('Test loss:', score[0])
# print('Test accuracy:', score[1])
# model.save_weights('model_weights.h5')
```
